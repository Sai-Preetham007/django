# **Web Scrapping**

```python
# ! pip install bs4
```

```python
import bs4
import requests
```

```python
url = "https://pin.it/1FsxOAsQH"
```

```python
res = requests.get(url)
```

```python
print(res.content.decode("utf-8"))
```

```python
soup = bs4.BeautifulSoup(res.content)
```

```python
soup.find_all("img")
```

```python
soup.find_all("img")[0]
```

```python
soup.find_all("img")[0].attrs
```

```python
soup.find_all("img")[0].attrs["src"]
```

```python
with open("Zhao Lusi.jpg", "wb") as file:
    img_url = soup.find_all("img")[0].attrs["src"]
    res = requests.get(img_url)
    
    file.write(res.content)
```

# **Working with Images**

```python
# ! pip install opencv-python
```

```python
import cv2
import matplotlib.pyplot as plt
```

## Reading the image file

```python
img = cv2.imread("Zhao Lusi.jpg")
```

## Using matplotlib to display the image

```python
plt.imshow(img)   # if you are using matplotlib with imshow it will be in BGR format. ---> try using cv2.imshow(img) then no need to do any color conversions.
plt.show()

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)        # cv2.cvtColor ---> convert color,  cv2.COLOR_BGR2RGB ---> BGR to RGB.

plt.imshow(rgb_img)
plt.axis("off")                                       # turning off axis
plt.show()
```

## Cropping the image

```python
plt.imshow(rgb_img)
plt.show()

plt.imshow(rgb_img[100:750, 50:600])            # [rows, col] ---> (rows = height = y_axis), (columns = width = x_axis)
plt.axis("off")
plt.show()
```

## Using cv2 to display the image
* It will display the image externally

```python
cv2.imshow("Zhao Lusi - Actress", img)
```

```python
img = cv2.imread("Zhao Lusi.jpg")

cv2.imshow("Zhao Lusi - Actress", img)          # it will not show the image in vscode, it will display the image externally. ---> go see the taskbar.

# cv2.waitKey(0)              # waitKey = 0 ---> wait for infinite time.   ---> no need to use detsroyAllWindows()

cv2.waitKey(5000)              # let say waitKey = 5000 ---> wait for 5 sec.
cv2.destroyAllWindows()        # after waiting for 5 sec's it will destroy the image.
```

## Converting to greyscale image

```python
grey_img = cv2.imread("Zhao Lusi.jpg", cv2.IMREAD_GRAYSCALE)
cv2.imshow("Grey Image", grey_img)

cv2.waitKey(3000)
cv2.destroyAllWindows()
```
