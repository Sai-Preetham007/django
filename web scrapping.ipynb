# **Scrapping Images from a website**
* Beautiful Soup
* Scrapy

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.passiton.com/inspirational-quotes?page=2"

response = requests.get(url)
response.content
```

```python
soup = BeautifulSoup(response.content)

soup
```

```python
soup.find_all("img")
```

```python
images = [i for i in soup.find_all("img") if ".jpg" in i.attrs["src"]]

images
```

```python
for i,j in enumerate(images):
    with open(f"inspirational_quotes/{i+1}.jpg", "wb") as file:
        img_url = j.attrs["src"]
        response = requests.get(img_url)

        file.write(response.content)

# Go and check inspirational quotes folder. ---> we have successfully scrapped the images from a website.
```
